```python
import numpy as np
import pickle
from softmax import SoftmaxClassifier
from solver import Solver
from sklearn.model_selection import train_test_split
```

```python
pickle_in = open("mnist.pkl","rb")
data_in = pickle.load(pickle_in, encoding='latin1')
```

```python
data_train = data_in['training_images']
labels_train  = data_in['training_labels']
X_train,X_val,y_train,y_val = train_test_split(data_train, labels_train, test_size=0.1)
```

```python
data = {
    'X_train': X_train,
    'y_train': y_train,
    'X_val': X_val,
    'y_val': y_val
}
model = SoftmaxClassifier(input_dim=784,hidden_dim=None,reg=0)
solver = Solver(
        model,
        data,
        update_rule='sgd',
        optim_config={ 'learning_rate': 0.0003 },
        lr_decay=1,
        num_epochs=100,
        batch_size=1000,
        print_every=1000
    )
val_accuracy = solver.train()
```
