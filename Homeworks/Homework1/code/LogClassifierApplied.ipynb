```python
import numpy as np
import pickle
from logistic import LogisticClassifier
from solver import Solver
```

```python
pickle_in = open("data.pkl","rb")
data_in = pickle.load(pickle_in, encoding='latin1')
```

```python
# Train-test split
X_train = data_in[0][:500]
X_val = data_in[0][500:750]
X_test = data_in[0][750:]
y_train = data_in[1][:500]
y_val = data_in[1][500:750]
y_test = data_in[1][750:]
```

```python
data = {
    'X_train': X_train,
    'y_train': y_train,
    'X_val': X_val,
    'y_val': y_val
}
model = LogisticClassifier(input_dim=20,weight_scale=1)
solver = Solver(
        model,
        data,
        update_rule='sgd',
        optim_config={ 'learning_rate': 0.9 },
        lr_decay=1,
        num_epochs=1000,
        batch_size=100,
        print_every=1000
    )
val_accuracy = solver.train()
```

```python
data = {
    'X_train': X_train,
    'y_train': y_train,
    'X_val': X_val,
    'y_val': y_val
}
hidden_dims = [5,10,20,30]
regs = [0.0001,0.0003,0.001,0.003,0.01,0.03]
results = []
for dim in hidden_dims:
    results.append([])
    for r in regs:
        model = LogisticClassifier(input_dim=20,hidden_dim=dim,reg=r,weight_scale=1)
        solver = Solver(
                model,
                data,
                update_rule='sgd',
                optim_config={ 'learning_rate': 0.9 },
                lr_decay=1,
                num_epochs=1000,
                batch_size=100,
                print_every=1000
            )
        val_accuracy = solver.train()
        results[-1].append(val_accuracy)
```
